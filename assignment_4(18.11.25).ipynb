###### [1번 문제]
파일 s.txt 내의 다음 내용과 함께 아래 문제의 코드를 제시하시오.

> pig ham  
> cat dog  
> ham bird  
> dog pig

    - 1) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다.

```python
import os

list=[]
f = open('s.txt', 'r')                     #파일 읽기
line = f.readlines()                       #한줄씩 불러와서 리스트에 저장
for i in range(4):
    list.append(line[i].rstrip('\n'))      #반복해서 개행문자 제거
l = sorted(list)                           #정렬
f1 = open('s1.txt', 'w')
for i in range(4):
    if(i==3):
        f1.write(l[i])                     #마지막 줄 개행문자 없음
    else:
        f1.write(l[i]+'\n')                #리스트를 저장하면서 줄바꿈
f2 = open('s1.txt', 'r')                   #파일 출력
print(f2.read())
f2.close()
f1.close()
f.close()
```

    - 2) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.

```python
list = []
list1=[]
list2=[]
f = open('s.txt', 'r')
line = f.readlines()                     #한줄씩 불러와 리스트에 저장

for i in range(4):
    list.append(line[i].rstrip('\n'))    #개행문자 제거 후 리스트에 저장

for i in range(4):
    list1.append(line[i][4])             #각 리스트의 인덱스의 5번째 요소로 정렬을 해야하므로 그 요소들만 따로 불러 리스트에 저장
list1.sort()                             #리스트 정렬   

for i in range(4):
    for j in range(4):
        if(list1[i]==list[j][4]):        #5번째 요소와 일치하는지 검사
            list2.append(list[j])        #일치하는 순서대로 새 리스트에 저장
            
f1 = open('s2.txt', 'w')
for i in range(4):
    if(i==3):
        f1.write(list2[i])               #마지막 줄은 개행문자없이 저장
    else:
        f1.write(list2[i]+'\n')          #나머지 줄은 개행문자 포함 저장

f.close()
f1.close()

with open('s2.txt', 'r') as f2:
    print(f2.read())
```

```python
import hashlib

f = open('access.txt', 'w')
f.close()

while '1':
    str = """Welcome to Our Service
    1. Sign Up
    2. Sign In
    3. Quit"""
    print(str)
    user=input()
    
    if user == '1':
        print('ID : ', end = "")
        ID = input()
        
        with open('access.txt', 'r') as f2:
            text = f2.read()
            
            if text!='':                                #텍스트가 비어있지 않을때
                while text.find(':') != -1:             #':'찾는 구문
                    start = text.find(':')
                    end = text.find('\n')
                    text = text[:start]+text[end+1:]
                comp=text.split()                       #ID만 따로 저장된 문자열을 공백으로 쪼개어 리스트로 저장
                i=0
                while '1':                    
                    if i>=len(comp):                    #반복하되 ID개수보다 커지면 종료
                        break;
                    if comp[i]=="[" + ID + "]":         #ID중복이면 다시 입력
                        print("Sorry, the entered ID is already used.")
                        print('ID : ', end = "")
                        ID = input()
                        i=0                             #다른 값을 입력 후 처음부터 중복있는지 다시 검색
                    elif comp[i]!="[" + ID + "]":   
                        i+=1
        print('Password : ', end = "")
        pw = input()
        print('Name : ', end = "")
        name = input()
        print('School : ', end = "")
        school = input()
    
        pw = hashlib.sha1(bytes(pw,'utf-8')).hexdigest() #암호화 과정

        info = "[" + ID + "] : " + "[" + pw + "], " + "[" + name + "], " + "[" + school + "]\n"
        with open('access.txt', 'a') as f1:
            f1.write(info)
    elif user == '3':
        break;
with open('access.txt', 'r') as f3:
    print(f3.read())
```

###### [3번 문제]
range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

 - 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
> range(stop)  
> range(start, stop[, step])

 - 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.
 - 3) frange 사용 예
   - 3-1) frange(0.5)
     - [0.0, 0.1, 0.2, 0.3, 0.4]
   - 3-2) frange(1.0, 2.0)
     - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
   - 3-3) frange(2.2, 4.0, 0.5)
     - [2.2, 2.7, 3.2, 3.7]

```python
def frange1(list):
    stop = float(list[0])                 #리스트의 0번째 인덱스를 저장
    result = []
    start = 0
    while start < stop:
        result.append(start)              #새 리스트에 저장
        start += 0.1
        start = round(start,1)            #소수점 첫번째 자리까지만
    return result
   
def frange2(list):
    if(len(list)==2):
        start = float(list[0])
        stop = float(list[1])
        result = []
        while start < stop:
            result.append(start)
            start += 0.1
            start = round(start,1)
        return result
    
    else:
        start = float(list[0])
        stop = float(list[1])
        step = float(list[2])
        result = []
        while start < stop:
            result.append(start)
            start += step
            start = round(start,1)
        return result
    
def frange(*a):                                  #가변인수로 받는다
    str1 = str(a)[1:-1]                          #튜플 제거 '(', ')'
    list = str1.split(',')                       #,기준으로 나눔
    if(list[-1]==''):                            #튜플의 요소가 1개였어도 ,가 있기 때문에 뒤에 공백이 포함된다 고로 공백 제거
        list.pop(1)
    if(len(list)==1):                            #길이가 1일때
        return frange1(list)
    else:                                        #길이가 2, 3일때
        return frange2(list)
    
print(frange(0.5))
print(frange(1.0, 2.0))
print(frange(2.2, 4.0, 0.5))
```

###### [4번 문제]  
가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오
 - sum() 사용 예
   - sum()
     - 0
   - sum(1, 2)
     - 3
   - sum(1, 2, 3, 4, 5)
     - 15
   - sum(1, 5, 7, 2, -10)
     - 5

```python
def sum(*a):
    result = 0
    str1 = str(a)[1:-1]
    list = str1.split(',')                       #,기준으로 나눔
    if(list[0]==''):                             #함수의 요소가 없을때 바로 반환
        return 0
    elif(list[-1]==''):                          #튜플의 요소가 1개였어도 ,가 있기 때문에 뒤에 공백이 포함된다 고로 공백 제거
        list.pop(1)
    for i in range(len(list)):                   #요소의 개수만큼 반복
        result += int(list[i])
    return result
  
print(sum())
print(sum(1,2))
print(sum(1,2,3,4,5))
print(sum(1,5,7,2,-10))
```

###### [5번 문제]
여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.
 - 1) 다음에 제시되는 함수들을 모두 이용해야 한다.
   - split
   - map
   - join
 - 2) myinitial() 함수 사용 예
   - myinitial("as soon as possible")
     - ASAP

```python
def f(x):
    return x[0].upper()              #각 원소의 첫번째 문자를 대문자로 반환

def myinitial(str):
    list = str.split()                #공백으로 분리
    m = map(f, list)
    str1 = ''.join(m)
    return str1

print(myinitial("as soon as possible"))
```

###### [6번 문제]
음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.
 - [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n):
    if(n==1 or n==0): # n이 1이나 0일땐 1반환
        return 1
    return n*myfact(n-1) #n*(n-1)!,(n-1)*(n-2)! 이런방법으로 n이 1될때까지 재귀적으로 돈다.

print(myfact(2965))            #내 컴퓨터에서 돌아가는 최대 크기
# 이 이상은 'maximum recursion depth exceeded in comparison' 이란 문구가 뜨며 오류
```

```python
def HW_Result2(ans):      #불용어 제거 후 사전으로 만들어주는 함수
    for i in range(len(ans)):
        for j in range(len(stop_word_en)): #영문 불용어 제거 과정
            if ans[i].lower() == stop_word_en[j]:
                ans[i] = ''
    ans = ' '.join(ans)
    ans = ans.split()
    for i in range(len(ans)):
        for j in range(len(stop_word_kr)):
            key = ans[i].find(stop_word_kr[j])     #ans[i] 에서 한글 불용어를 찾아 key에 저장
            if key!=-1:                          #key가 -1 이 아니면 ans[i]에 한글불용어가 있다는 것
                if ans[i][key]==ans[i][-1]:          #ans[i]에서 찾아낸 한글 불용어가 맨 끝단어이면
                    ans[i] = ans[i][:len(ans[i])-1]
                    #그 단어 제거 ex)로그인이 에서 '이'가 맨 끝단어이며 한글 불용어이기 때문에 제거
                    break #제거 된 후 계속 반복문이 실행되면 이후 갱신된 다른 글자가 제거될수 있기 때문에 중단
                if len(stop_word_kr[j])!=1:      #발견된 한글 불용어의 길이가 1이 아니면
                    if ans[i][key+len(stop_word_kr[j])-1]==ans[i][-1]:#ans[i]에서 한글 불용어 뒤로 글자가 없으면
                        ans[i] = ans[i][:len(ans[i])-len(stop_word_kr[j])] #ans[i]에서 한글 불용어 길이 만큼 뒤에서 제거
                        break
    ans = ' '.join(ans)
    ans = ans.split()
    d={}
    for x in range(len(ans)):
        if ans[x] in d:                    #사전에 키가 있다면 if문
            d[ans[x]] += 1
        else:
            d[ans[x]]=1                    #없다면 else문 실행
    return d
```
