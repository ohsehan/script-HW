###### [3번 문제]
다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)
 - 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
    - count: Counter 인스턴스가 지니는 초기 정수 값
    - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)  
    
   >     >>> c = Counter(10)  
   >     >>> d = Counter(10, 2) 

 - 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

   >     >>> print(c)  
   >     [Count (step: 1)] 10  
   >     >>> print(d)  
   >     [Count (step: 2)] 10

 - 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

   >     >>> c.incr()  
   >     >>> d.incr()  
   >     >>> print(c)  
   >     [Count (step: 1)] 11  
   >     >>> print(d)  
   >     [Count (step: 2)] 12 

 - 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

   >     >>> c()  
   >     >>> d()  
   >     >>> print(c)  
   >     [Count (step: 1)] 12  
   >     >>> print(d)  
   >     [Count (step: 2)] 14

 - 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

   >     >>> c = c + 5  
   >     >>> d = d - 5  
   >     >>> print(c)  
   >     [Count (step: 1)] 17  
   >     >>> print(d)  
   >     [Count (step: 2)] 9 

 - 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

   >     >>> print(c) > 10  
   >     True  
   >     >>> print(d) > 10  
   >     False  
   >     >>> print(c) < 10  
   >     False  
   >     >>> print(d) < 10  
   >     True  
   >     >>> print(c) == 17  
   >     True  
   >     >>> print(d) != 9  
   >     False 

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step

    def __str__(self):
        return "[Count(step : %s)] %s"%(self.step, self.count)
    
    def incr(self):
        self.count += self.step
    
    def __call__(self): #인스턴스 객체 직접 호출 메소드
        self.incr()     #내부에서 일어나는 일이므로 self사용
    
    def __add__(self, num):#더하기 연산 메소드
        self.count += num
        return Counter(self.count, self.step)#메소드에 다시 저장
    
    def __sub__(self, num):
        self.count -= num
        return Counter(self.count, self.step)
        
    def __gt__(self, num):# '>'에 대한 메소드
        if self.count>num:
            return True
        else:
            return False
        
    def __lt__(self, num):# '<'에 대한 메소드
        if self.count<num:
            return True
        else:
            return False
        
    def __eq__(self, num):# '=='에 대한 메소드
        if self.count==num:
            return True
        else:
            return False
        
    def __ne__(self, num):# '!='에 대한 메소드
        if self.count!=num:
            return True
        else:
            return False
        
c = Counter(10) 
d = Counter(10, 2) 
print(c)
print(d) 
print()

c.incr() 
d.incr()
print(c) 
print(d) 
print()

c() 
d()
print(c) 
print(d) 
print()

c = c + 5 
d = d - 5 
print(c) 
print(d) 
print()

print(c > 10)
print(d > 10)
print(c < 10) 
print(d < 10) 
print(c == 17)
print(d != 9)
```

###### [4번 문제]
다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
 - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l): # MySet 생성자
        for e in l: #l안의 원소를 다 돌때까지 반복
            self.append(e) # 자기 자신에 list로 추가
        MySet.eliminate_duplicate(self)#메소드 실행
    
    def __str__(self):#인스턴스를 출력해주는 메소드
        result = "MySet: {"#result안에 문자열 저장
        for e in self:#자신에게 있는것 다 돌때까지 반복
            result = result + str(e) + " ,"#하나하나씩 result에 문자열로 저장
        result = result[0:len(result)-2] + "}"#result에 있는 마지막','대신 '}' 저장
        return result #result 반환

    @staticmethod    #정적메소드
    def eliminate_duplicate(l):#인스턴스 객체와 상관없이 클래스 이름 공간에 존재하는 메소드
                               #클래스 이름을 이용하여 직접 호출 가능한 메소드
        s = [] #리스트 s 만듬
        for e in l: #l안의 원소를 다 돌때까지 반복
            if e not in s:#리스트 s안에 없으면 아래문장 실행
                s.append(e) #s에 추가
        l[:] = [] # l을 비운다
        for e in s:#s안의 원소를 다 돌때까지 반복
            l.append(e)#l에 추가

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

###### [5번 문제]
4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

 - 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
 - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

   >     >>> u = s | t 
   >     >>> print(u) 
   >     MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 

 - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

   >     >>> u = s & t 
   >     >>> print(u) 
   >     MySet: {2, 3} 

 - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

   >     >>> s = MySet([1, 2, 3]) 
   >     >>> t = MySet([3, 4, 5]) 
   >     >>> u = s - t 
   >     >>> print(u) 
   >     MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l): # MySet 생성자
        for e in l: #l안의 원소를 다 돌때까지 반복
            self.append(e) # 자기 자신에 list로 추가
        MySet.eliminate_duplicate(self)#메소드 실행
    
    def __str__(self):#인스턴스를 출력해주는 메소드
        result = "MySet: {"#result안에 문자열 저장
        for e in self:#자신에게 있는것 다 돌때까지 반복
            result = result + str(e) + " ,"#하나하나씩 result에 문자열로 저장
        result = result[0:len(result)-2] + "}"#result에 있는 마지막','대신 '}' 저장
        return result #result 반환
    
    def __or__(self, l): #'|'에 대한 메소드
        tmp=[] #빈리스트 생성
        for e in l:
            tmp.append(e)#l에 있는원소들 tmp에 추가
        for e in self:
            tmp.append(e)#자신에게 있는 원소들 tmp에 추가
        MySet.eliminate_duplicate(tmp)#중복제거
        return MySet(tmp)
    
    
    def __and__(self, l):
        tmp=[] #빈리스트 생성
        tmp2=[]#빈리스트 생성
        for e in l:
            tmp.append(e)#l에 있는원소들 tmp에 추가
        for e in self:
            if e in tmp:
                tmp2.append(e)#self와 tmp에 둘다 있으면 tmp2에 저장
        tmp2.sort()#정렬
        return MySet(tmp2)
    
    def __sub__(self, l):
        tmp=[] #빈리스트 생성
        tmp2=[]#빈리스트 생성
        for e in self:
            tmp.append(e)#l에 있는원소들 tmp에 추가
        for e in l:
            if e in tmp:
                tmp2.append(e)#self와 tmp에 둘다 있으면 tmp2에 저장, 교집합한 과정
        tmp[:]=[] #리스트 비우기
        for e in self: #self의 원소중 tmp2에 없으면 tmp에 추가
            if e not in tmp2:
                tmp.append(e)
        return MySet(tmp)
    
    @staticmethod    #정적메소드
    def eliminate_duplicate(l):#인스턴스 객체와 상관없이 클래스 이름 공간에 존재하는 메소드
                               #클래스 이름을 이용하여 직접 호출 가능한 메소드
        s = [] #리스트 s 만듬
        for e in l: #l안의 원소를 다 돌때까지 반복
            if e not in s:#리스트 s안에 없으면 아래문장 실행
                s.append(e) #s에 추가
        l[:] = [] # l을 비운다
        for e in s:#s안의 원소를 다 돌때까지 반복
            l.append(e)#l에 추가
        l.sort() #원소 정렬

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s | t 
    print(u) 
    u = s & t 
    print(u) 
    u = s - t
    print(u) 
```

###### [8번 문제]
위 7번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

 - 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
 - getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.
 

    - reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
    - reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
   >     >>> w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
   >     >>> w4.getWordsFrequency()
   >     [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]
   >     >>> w4.getWordsFrequency(reverse=true)
   >     [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def __init__(self, *l):
        self.url = list(l)
            
    def getWordsFrequency(self, reverse = False):
        self.list=[]
        d = SearchEngine.getWordsFrequency(self)
        D = {}
        key = list(d.keys())              #키값을 받아 리스트에 저장
        value = list(d.values())          #벨류값을 받아 리스트에 저장
        if reverse == True:
            value.sort()                      #벨류값 정렬
            for i in value:
                for j in key:
                    if(d[j] == i):            #사전의 벨류와 정렬한 벨류가 같다면
                        D[j] = i              #새로운 사전에 저장
            self.list = list(D.items())              #튜플로 키,값 저장
            
        else:
            value.sort(reverse=True)          #벨류값 거꾸로 정렬
        
            for i in value:
                for j in key:
                    if(d[j] == i):            #사전의 벨류와 정렬한 벨류가 같다면
                        D[j] = i              #새로운 사전에 저장
            self.list = list(D.items())               #튜플로 키,값 저장 
        return self.list
    
    def __iter__(self):
        self.num = 0
        return self

    def __next__(self):
        if self.num >=len(self.list): 
            raise StopIteration
        str1 = self.list[self.num]
        self.num += 1
        return str1
                      
w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
print(w4.getWordsFrequency())
print(w4.getWordsFrequency(reverse=True))
```
